```python
df
```

There are several problems.  Firstly, Pandas has added its own index to the table (the numbers at the left).  You can explicitly add your own by declaring an existing column as the index.

```python
df = pd.read_csv("data/celebrations.csv", index_col=0)
df
```

Better, but the 'header' row is a problem still. We can declare no headers (the top row that gives columns names) by overriding the default using header=None.

```python
df = pd.read_csv("data/celebrations.csv", index_col=0, header=None)
df
```

There's a problem with the data.  Twix contains an entry of 25, when the max should be 8.  



```python
df.at['TWIX',13] = 2
```

```python
df
```

```python
df.at['TWIX',14] = 5
```

```python
df
```

Now add a column called 'mean' to the df, with the values computed as below.

```python
df['mean'] = df.mean(axis=1)
df
```

Now sort values by the mean column.

```python
df.sort_values('mean')
```
